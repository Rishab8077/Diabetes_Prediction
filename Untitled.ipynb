```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from statsmodels.graphics.gofplots import qqplot
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.svm import SVC
import pickle
```

```python
df= pd.read_csv('diabetes.csv')
df.head()
```

```python
#feature like gulcose, skinthickness, BMI & age cannot have value 0.
features=['Glucose','BloodPressure','SkinThickness','BMI','Age']
null_feat=[ft for ft in features if df[ft].unique().all()== 0]
```

```python
null_feat
```

```python
for feat in null_feat:
    df[feat].replace(0,np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
for feature in null_feat:
    median=df[feature].median()
    df[feature].fillna(median,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
X=df.drop(['Outcome'],axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
random_forest_model = RandomForestClassifier(random_state=4219)
random_forest_model.fit(X_train, y_train)
```

```python
y_pred1 = random_forest_model.predict(X_test)
print(confusion_matrix(y_test,y_pred1))
print(accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
logistic_regression=LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

```python
y_pred3 = logistic_regression.predict(X_test)
print(confusion_matrix(y_test,y_pred3))
print(accuracy_score(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

```python
from sklearn.model_selection import RandomizedSearchCV
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
params = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = params,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_estimator_
```

```python
random_forest=RandomForestClassifier(max_depth=25, max_features='sqrt', n_estimators=1000, random_state=4219)
random_forest.fit(X_train,y_train)
```

```python
y_pred4 = random_forest.predict(X_test)
print(confusion_matrix(y_test,y_pred4))
print(accuracy_score(y_test,y_pred4))
print(classification_report(y_test,y_pred4))
```

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_pred5 = xgb.predict(X_test)
print(confusion_matrix(y_test,y_pred5))
print(accuracy_score(y_test,y_pred5))
print(classification_report(y_test,y_pred5))
```

```python
svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_pred6 = svc.predict(X_test)
print(confusion_matrix(y_test,y_pred6))
print(accuracy_score(y_test,y_pred6))
print(classification_report(y_test,y_pred6))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
y_pred7 = knn.predict(X_test)
print(confusion_matrix(y_test,y_pred7))
print(accuracy_score(y_test,y_pred7))
print(classification_report(y_test,y_pred7))
```
